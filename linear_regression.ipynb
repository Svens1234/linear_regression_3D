```python
from sklearn.linear_model import LinearRegression
```

```python
X = houses[["dim_1"]]
y = houses["price"]
reg = LinearRegression().fit(X.values.reshape(-1, 1),y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
[b] = reg.coef_
a = reg.intercept_
def reg_prediction(dim_1):
    return a + b * dim_1
```

```python
reg_prediction(X.dim_1[0])
```

```python
reg.predict(X[0:1])[0]
```

```python
fig=plt.figure(figsize=(10,10))#figsize-размер точек 
ax=plt.axes()
ax.scatter(houses["dim_1"], houses["price"],s=100)#s-размер графика
ax.plot([X.dim_1.min(), X.dim_1.max()], [reg_prediction(X.dim_1.min()), reg_prediction(X.dim_1.max())], c="red")
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
ax=plt.axes(projection="3d")
ax.scatter(houses["dim_1"], houses["price"], houses["price"], s=100)
ax.set_xlabel("dim_1")
ax.set_ylabel("dim_2")
ax.set_zlabel("price")
plt.show()
```
